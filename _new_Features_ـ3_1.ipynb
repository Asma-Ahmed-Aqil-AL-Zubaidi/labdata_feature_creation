# 1-Blood Pressure Features

# Pulse Pressure

```python
df3["Pulse_Pressure"] = df3["Systolic Pressure"] - df3["Diastolic Pressure"]
```

# Mean Arterial Pressure (MAP)

```python
df3["MAP"] = (df3["Systolic Pressure"] + 2 * df3["Diastolic Pressure"]) / 3
```

# BP Ratio

```python
df3["BP_Ratio"] = df3["Systolic Pressure"] / df3["Diastolic Pressure"]
```

# 2-Kidney Function Features

```python
df3["Urea_Creatinine_Ratio"] = df3["Urea in Serum"] / df3["Creatinine in Serum"]
```

```python
df3["BUN_eGFR_Ratio"] = df3["Blood Urea Nitrogen (BUN)"] / df3["Estimated Glomerular Filtration Rate(eGFR)"]
```

# 3. Lipid Profile Features

```python
df3["Non_HDL_Cholesterol"] = df3["Cholesterol"] - df3["HDL Cholesterol"]
```
